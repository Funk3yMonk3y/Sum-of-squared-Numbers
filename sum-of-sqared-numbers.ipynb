```python
from tabulate import tabulate
```

```python
k = 0
a =(k*k)
n = 30
l = []
m = []
o = []
p = []
#Index 0 to index 30 square, emptylists
while k < n: 
   k = k + 1
   a =k*k
   b = ((k*(k+1)*((2*k)+1))/6)
   l.append(a)
   m.append(k)
   o.append(b)
   p.append(sum(l))
#Formula loop, list item adds
   
table = [[[m]],[o],[p]]
#Tablelayout
print(tabulate(table,headers = "firstrow"))
```
